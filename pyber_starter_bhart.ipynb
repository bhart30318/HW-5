```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# File to Load (Remember to change these)
cdata_path = "Data/city_data.csv"
rdata_path = "Data/ride_data.csv"

cdata_df = pd.read_csv(cdata_path)
rdata_df = pd.read_csv(rdata_path)
```

```python
Rgroup = rdata_df.groupby(['city']+['date']+['fare']+['ride_id'])
a = int(len(Rgroup))
b = int(len(rdata_df))

if a == b:
    print('no need to consolidate Ridedata')
else:
    rdata_df = rdata_df.drop.duplicates()
    
c = cdata_df['city'].nunique()
d = cdata_df['city'].count()

if c == d:
    print('no need to consolidate citydata')
else:
    cgroup = cdata_df.groupby(['city']+['type'])
    cdata_df = cgroup['driver_count'].sum()
    cdata_df = cdata_df.reset_index()
    e = cdata_df['city'].count()
    print ('citydata consolidated from '+str(d)+' to ' +str(e))
```

```python
combined_df = cdata_df.merge(rdata_df, how='left', on='city')
combined_df.to_csv("data/combined.csv", index=False)

# Read the City and Ride Data

# File to Load (Remember to Change These)

# Read Purchasing File and store into Pandas data frame

# Combine the data into a single dataset

# Display the data table for preview
combined_df.head()
```

## Total Fares by City Type

```python
# Calculate Type Percents



# Build Pie Chart

farest = combined_df.groupby(['type'])['fare'].sum()
labels = ["Urban", "Rural", "Suburban"]
colors = ["red", "yellow", "blue"]
explode = (0,0,.1)

# Build Pie Chart

plt.title("% of Total Fares by City")
plt.pie(farest, explode=explode, labels=labels, colors=colors, autopct ="%1.1f%%", shadow = True, startangle=80)
plt.axis("equal")

# Save Figure

plt.savefig("Pyber Pie Chart Total Fares By City")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents

ridest = combined_df.groupby(['type']).count()['ride_id']
labels = ["Urban", "Rural", "Suburban"]
colors = ["red", "yellow", "blue"]
explode = (0,0,.1)

# Build Pie Chart

plt.title("% of Total Rides by City")
plt.pie(ridest, explode=explode, labels=labels, colors=colors, autopct ="%1.1f%%", shadow = True, startangle=80)
plt.axis("equal")


# Save Figure

plt.savefig("Pyber Pie Chart Total Rides By City")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents



# Build Pie Charts

drivert = combined_df.groupby(['type']).agg({'driver_count':['sum']})
labels = ["Urban", "Rural", "Suburban"]
colors = ["red", "yellow", "blue"]
explode = (0,0,.1)

# Build Pie Chart

plt.title("% of Total Rides by City")
plt.pie(drivert, explode=explode, labels=labels, colors=colors, autopct ="%1.1f%%", shadow = True, startangle=80)
plt.axis("equal")

# Save Figure

plt.savefig("Pyber Pie Chart Total Drivers By City")
```

```python
# Show Figure
plt.show()
```
